### 3-4．다시 데이터 준비로
#### 데이터 리드

```python
# Pandas의 로드
import pandas as pd

# 파일 리드
d_df = pd.read_csv('data.csv', sep=',')

# 선두에서 5행까지 표시
d_df.head()
```

```python
# 데이터의 건수, 항목수를 확인
print(d_df.shape)

# 데이터 형을 확인
print(d_df.dtypes)
```

```python
print(d_df.isnull().sum(axis=0))
```

### 결손값 제외

```python
# job、education에 결손값이 포함되어 있는 행을 삭제
d_df = d_df.dropna(subset=['food', 'member'])

# 데이터의 건수, 항목수를 확인
print(d_df.shape)
```

```python
print(d_df.isnull().sum(axis=0))
```

### 결손값을 보완

```python
# 결손값을 「unknown」으로 치환
d_df = d_df.fillna({'destination':'unknown'})

# 선두에서 5행까지 표시
d_df.head()
```

```python
print(d_df.isnull().sum(axis=0))

print(d_df.shape)
```

### 특이값(이상치)의 제외

```python
# age가 18세 이상, 100세 미만이 아닌행을 제외
d_df = d_df[d_df['age'] >= 18]
d_df = d_df[d_df['age'] < 100]

# 데이터의 건수, 항목수를 확인
print(d_df.shape)
```

### 문자열값을 통합

```python
d_df.loc[(d_df['job'] == 'auther') |
         (d_df['job'] == 'baker') |
         (d_df['job'] == 'bank clerk') |
         (d_df['job'] == 'chef') |
         (d_df['job'] == 'dentist') |
         (d_df['job'] == 'designer') |
         (d_df['job'] == 'doctor') |
         (d_df['job'] == 'driver') |
         (d_df['job'] == 'engineer') |
         (d_df['job'] == 'farmer') |
         (d_df['job'] == 'firefighter') |
         (d_df['job'] == 'fisherman') |
         (d_df['job'] == 'lawyer') |
         (d_df['job'] == 'model') |
         (d_df['job'] == 'nurse'), 'job2'] = 'worker'

# 선두에서 5행까지 표시
d_df.head()
```

```python
# month가 jan、feb、mar인 데이터를 1Q로 치환
d_df.loc[(d_df['season'] == 'fall') |
         (d_df['season'] == 'autumn'), 'season'] = 'fall'

d_df.head(5)
```

```python
# 연습문제11

d_df.loc[(d_df['prev_destination'] == 'Hawaii of USA'), 'pre_d'] = 'hawaii'

d_df.loc[(d_df['prev_destination'] != 'Hawaii of USA'), 'pre_d'] = 'Not_hawaii'


# 선두에서 5행까지 표시
d_df.head(5)
```

### 수치의 통합

```python
# duration이 300미만이면 short로 치환
d_df.loc[d_df['number_yoy'] < 25, 'number_yoy2'] = 'short'

# duration이 300이상이면 long으로 치환
d_df.loc[d_df['number_yoy'] >= 25, 'number_yoy2'] = 'long'

# 선두에서 5행까지 표시
d_df.head()
```

```python
# duration이 300미만이면 short로 치환
d_df.loc[d_df['member'] < 4, 'member2'] = 'few'

# duration이 300이상이면 long으로 치환
d_df.loc[d_df['member'] >= 4, 'member2'] = 'many'

# 선두에서 5행까지 표시
d_df.head()
```

```python
# duration이 300미만이면 short로 치환
d_df.loc[d_df['period'] < 8, 'period2'] = 'short'

# duration이 300이상이면 long으로 치환
d_df.loc[d_df['period'] >= 8, 'period2'] = 'long'

# 선두에서 5행까지 표시
d_df.head()
```

```python
# duration이 300미만이면 short로 치환
d_df.loc[d_df['budget'] < 500000, 'budget2'] = 'less'

# duration이 300이상이면 long으로 치환
d_df.loc[d_df['budget'] >= 500000, 'budget2'] = 'much'

# 선두에서 5행까지 표시
d_df.head(5)
```
