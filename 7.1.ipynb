### 4-1．고객의 특성을 알자
#### 데이터의 리드와 확인

```python
# Pandas의 로드
import pandas as pd

# 파일 리드
d_df = pd.read_csv('b-prep.csv', sep=',')

# 선두에서 5행까지 표시
d_df.head()
```

```python
# 데이터의 건수, 항목수를 확인
print(d_df.shape)

# 데이터 형을 확인
print(d_df.dtypes)
```

#### 데이터의 정규화

```python
# MinMaxScaler의 로드
from sklearn.preprocessing import MinMaxScaler

# 데이터 세트에서 목적변수를 제외
d_df = d_df.drop('pre_d', axis=1)

# 범위변환의 인스턴스를 생성
mc = MinMaxScaler()
mc.fit(d_df)

# d_df의 각항목을 정규화
d_df_mc = pd.DataFrame(mc.transform(d_df), columns=d_df.columns)
d_df_mc.head()
```

```python
# StandardScaler의 로드
from sklearn.preprocessing import StandardScaler

# Z변환 인스턴스 생성
sc = StandardScaler()
sc.fit(d_df)

# d_df의 각항목을 정규화
d_df_sc = pd.DataFrame(sc.transform(d_df), columns=d_df.columns)
d_df_sc.head()
```

```python
# 각 항목의 평균과 표준편차를 확인
print(d_df_sc.mean())
print(d_df_sc.std())
```
