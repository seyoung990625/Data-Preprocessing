```python
y_yes = d_df[d_df['y'] == 'y']
y_no = d_df[d_df['y'] == 'n']

y_age = [y_yes['age'], y_no['age']]

plt.boxplot(y_age)

plt.xlabel('y')
plt.ylabel('age')
ax = plt.gca()

plt.setp(ax, xticklabels = ['yes','no'])
plt.show()
```
